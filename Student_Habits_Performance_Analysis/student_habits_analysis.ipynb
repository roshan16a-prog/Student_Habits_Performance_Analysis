```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
gender_count = df['gender'].value_counts()
plt.pie(
    gender_count,
    labels=gender_count.index,
       wedgeprops={"edgecolor": "white"},
    colors = ["#b7c4a3", "#a3c9c7", "#e6b8a2"],
    autopct='%1.1f%%',
    startangle=90
)

plt.title('Gender Distribution')
plt.axis('equal')
plt.show()
```

```python
sns.set_theme(style="white")

plt.figure(figsize=(6, 6))

time_mean = df[
    ["study_hours_per_day", "sleep_hours", "screen_time"]
].mean()

plt.pie(
    time_mean,
    labels=["Study", "Sleep", "Screen"],
    autopct="%1.1f%%",
    startangle=140,
    colors=["#cdb4db", "#bde0fe", "#f4a6a6"],
    wedgeprops={"edgecolor": "white"},
    shadow=True
)

plt.title(
    "Average Daily Time Allocation of Students",
    fontsize=14,
    fontweight="bold",
    pad=10
)

plt.tight_layout()
plt.savefig(
    "plots/daily_time_allocation.png",
    dpi=300,
    bbox_inches="tight"
)
plt.show()
```

```python

df["screen_time"] = df["social_media_hours"] + df["netflix_hours"]

df["risk_flag"] = (
    (df["attendance_percentage"] < 75) |
    (df["mental_health_rating"] <= 2) |
    (df["screen_time"] > 4)
)
sns.set_theme(style="darkgrid")

plt.figure(figsize=(7, 5))

avg_scores = df.groupby("risk_group")["exam_score"].mean().reset_index()

sns.barplot(
    data=avg_scores,
    x="risk_group",
    y="exam_score",
    palette=["#f4a6a6", "#bde0fe"],
    
    edgecolor="black"
)

plt.title(
    "Academic Risk Group vs Average Exam Score",
    fontsize=14,
    fontweight="bold"
)

plt.xlabel("Student Risk Group")
plt.ylabel("Average Exam Score")

plt.tight_layout()
plt.savefig(
    "plots/academic_risk_vs_exam_score.png",
    dpi=300,
    bbox_inches="tight"
)
plt.show()
```

```python
df["screen_time"] = df["social_media_hours"] + df["netflix_hours"]

df["score_group"] = df["exam_score"].apply(
    lambda x: "High Scorers" if x >= 75 else "Low Scorers"
)

time_avg = df.groupby("score_group")[[
    "study_hours_per_day",
    "sleep_hours",
    "screen_time"
]].mean()
```

```python
 
sns.set_theme(style="whitegrid")

plt.figure(figsize=(8, 8))

plt.bar(
    time_avg.index,
    time_avg["study_hours_per_day"],
    label="Study",
    color="#cdb4db"   # light purple
)

plt.bar(
    time_avg.index,
    time_avg["sleep_hours"],
    bottom=time_avg["study_hours_per_day"],
    label="Sleep",
    color="#bde0fe"   # light blue
)

plt.bar(
    time_avg.index,
    time_avg["screen_time"],
    bottom=time_avg["study_hours_per_day"] + time_avg["sleep_hours"],
    label="Screen Time",
    color="#f4a6a6"   # light red
)

plt.title(
    "Time Allocation Trade-Off: High vs Low Scorers",
    fontsize=14,
    fontweight="bold"
)

plt.xlabel("Student Group")
plt.ylabel("Average Hours per Day")

plt.legend(title="Activity")
plt.tight_layout()
plt.savefig(
    "plots/Time_Allocation_Trade-Off: High_vs_Low_Scorers.png",
    dpi=300,
    bbox_inches="tight"
)
plt.show()
```

```python
sns.set_theme(style="whitegrid")

plt.figure(figsize=(9, 5))

sns.histplot(
    data=df,
    x="exam_score",
    hue="risk_group",
    bins=12,
    multiple="layer",
    kde=True,
    palette={
        "At Risk": "#f4a6a6",
        "Not At Risk": "#bde0fe"
    },
    edgecolor="white",
    alpha=0.85
)

plt.title(
    "Exam Score Distribution by Academic Risk Group",
    fontsize=14,
    fontweight="bold",
    pad=10
)

plt.xlabel("Exam Score")
plt.ylabel("Number of Students")

plt.tight_layout()
plt.savefig(
    "plots/exam_score_distribution_by_risk_group.png",
    dpi=300,
    bbox_inches="tight"
)
plt.show()
```

```python

sns.set_theme(style="white")

plt.figure(figsize=(10, 6))

pivot = df.pivot_table(
    values="exam_score",
    index="attendance_percentage",
    columns="study_hours_per_day",
    aggfunc="mean"
)

sns.heatmap(
    pivot,
    cmap="YlGnBu",
    linewidths=0.4,
    linecolor="white",
    cbar_kws={
        "label": "Average Exam Score",
        "shrink": 0.85
    }
)

plt.title(
    "Impact of Attendance and Study Hours on Exam Performance",
    fontsize=14,
    fontweight="bold",
    pad=12
)

plt.xlabel("Study Hours Per Day")
plt.ylabel("Attendance Percentage")

plt.tight_layout()
plt.savefig(
    "plots/attendance_study_vs_exam_score.png",
    dpi=300,
    bbox_inches="tight"
)
plt.show()
```
